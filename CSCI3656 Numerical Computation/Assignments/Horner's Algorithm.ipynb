```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
import math
myList = []
num = 1.92001
while num<2.08:
    myList.append(num)
    num += 0.00002
myVector = np.array(myList)
myList2 = []
def hornersalgorithm(poly, n, x):
    result = poly[0] 
    for i in range(1, n):
        result = result*x + poly[i]
    return result
polynomialList = [1,-18,144,-672,2016,-4032,5376,-4608,2304,-512]
polynomialArray = np.array(polynomialList)
for i in range(0,len(myVector)):
    myList2.append(hornersalgorithm(polynomialArray,len(polynomialArray),myVector[i]))
myVector2 = np.array(myList2)
plt.plot(myVector, myVector2, 'o', color='black');
```
