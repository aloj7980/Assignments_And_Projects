```python
import numpy as np
import math
import random
with open('mat1.txt', 'r') as f:
    m1 = [[float(num) for num in line.split(',')] for line in f]
with open('mat2.txt', 'r') as f:
    m2 = [[float(num) for num in line.split(',')] for line in f]
with open('mat3.txt', 'r') as f:
    m3 = [[float(num) for num in line.split(',')] for line in f]
with open('mat4.txt', 'r') as f:
    m4 = [[float(num) for num in line.split(',')] for line in f]
with open('mat5.txt', 'r') as f:
    m5 = [[float(num) for num in line.split(',')] for line in f]
print("%f %f %f %f %f %f %f %f %f %f"%(len(m1),len(m1[0]),len(m2),len(m2[0]),len(m3),len(m3[0]),len(m4),len(m4[0]),len(m5),len(m5[0])))
def countNonZero(a):
    count=0
    for i in range(0,len(a)):
        for j in range(0,len(a[i])):
            if a[i][j] != 0:
                count=count+1
    return count
print("%d %d %d %d %d"%(countNonZero(m1),countNonZero(m2),countNonZero(m3),countNonZero(m4),countNonZero(m5)))
def checkSymmetric(a):
    for i in range(0,len(a)):
        for j in range(0,len(a[i])):
            if a[i][j] != a[j][i]:
                return 0
    return 1
print("%d %d %d %d %d"%(checkSymmetric(m1),checkSymmetric(m2),checkSymmetric(m3),checkSymmetric(m4),checkSymmetric(m5)))
def checkDiagonal(a):
    for i in range(0,len(a)):
        for j in range(0,len(a[i])):
            if a[i][j] != 0 and i != j:
                return 0
    return 1
print("%d %d %d %d %d"%(checkDiagonal(m1),checkDiagonal(m2),checkDiagonal(m3),checkDiagonal(m4),checkDiagonal(m5)))
def checkOrthogonal(a):
    if np.linalg.det(a) == 0:
        return 0
    b=np.linalg.inv(a)
    for i in range(0,len(a)):
        for j in range(0,len(a[i])):
            if b[i][j] != a[j][i]:
                return 0
    return 1
print("%d %d %d %d %d"%(checkOrthogonal(m1),checkOrthogonal(m2),checkOrthogonal(m3),checkOrthogonal(m4),checkOrthogonal(m5)))
print("%d %d %d %d %d"%(np.linalg.matrix_rank(m1),np.linalg.matrix_rank(m2),np.linalg.matrix_rank(m3),np.linalg.matrix_rank(m4),np.linalg.matrix_rank(m5)))
def smallestSV(a):
    return np.linalg.svd(a)[1][len(np.linalg.svd(a)[1])-1]
def largestSV(a):
    return np.linalg.svd(a)[1][0]
print("%f %f %f %f %f"%(smallestSV(m1),smallestSV(m2),smallestSV(m3),smallestSV(m4),smallestSV(m5)))
print("%f %f %f %f %f"%(largestSV(m1),largestSV(m2),largestSV(m3),largestSV(m4),largestSV(m5)))
print("%f %f %f %f %f"%(largestSV(m1)/smallestSV(m1),largestSV(m2)/smallestSV(m2),float('inf'),1,largestSV(m5)/smallestSV(m5)))
```
